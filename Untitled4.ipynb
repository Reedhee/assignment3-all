```python
driver.get("https://www.youtube.com/")
```

```python
video=driver.find_element(By.XPATH,'//input[@class="ytd-searchbox"]')
video.send_keys("adipurush")
```

```python
search=driver.find_element(By.XPATH,'//button[@class="style-scope ytd-searchbox"]')
search.click()
time.sleep(2)
```

```python
for _ in range(20):
    driver.execute_script("window.scrollBy(0,500)")
comment_elems = driver.find_elements(By.XPATH,'//yt-formatted-string[@id="content-text"]')
upvote_elems = driver.find_elements(By.XPATH,'//span[@id="vote-count-left"]')
time_elems = driver.find_elements(By.XPATH,'//yt-formatted-string[@class="published-time-text above-comment style-scope ytd-comment-renderer"]')

for comment, upvote, time_elem in zip(comment_elems, upvote_elems, time_elems):
    try:
        comment_text = comment.text
        upvote_count = upvote.text
        time_posted = time_elem.get_attribute('title')
        print("Comment:", comment_text)
        print("Upvotes:", upvote_count)
        print("Time Posted:", time_posted)
        comment_count += 1
        if comment_count >= 500:
            break

    except NoSuchElementException:
        continue
```
